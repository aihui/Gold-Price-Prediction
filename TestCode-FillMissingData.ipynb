```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore') #this would remove any deprecated warning
```

```python
# Read in csv file
dateparser = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d')
data = pd.read_csv('data/gld-small-dataset.csv', parse_dates=['Date'], date_parser=dateparser)


cols = data.columns #store original column names
data.set_index("Date", inplace=True)
data.head(10)
```

```python
print(data.columns)
```

```python
data = data.resample("D").ffill().reset_index()
```

```python
data
```

```python
# Sort by date in ascending order
data = data[cols] #revert column order
data.sort_values(by='Date', inplace=True)
data.to_csv('data/gld-small-dataset-nomissing.csv', index=False)
```
